```python
import pandas as pd #판다스 가져오기
from pandas import Series, DataFrame #판다스에서 데이터프레임, 시리즈 가져오기
import numpy as np #넘파이 가져오기
%matplotlib notebook 
import matplotlib.pyplot as plt #파이플롯 가져오기
from sklearn.preprocessing import PolynomialFeatures #사이킷런에서 다항회귀 가져오기
from sklearn.linear_model import LinearRegression #선형회귀 가져오기
import matplotlib as mpl #매트플롯라이브러리 가져오기
plt.rcParams['axes.unicode_minus'] = False #마이너스 기호 깨짐 방지 코드
plt.rcParams['font.family'] = 'APPLEGothic' #한글 폰트 설정
df=pd.read_csv('/Users/yanghaein/pythongit/master.csv') #자료 가져오기
#print(df)
#print(df.describe())
#print(df.info())
df1=df.dropna() #결측치 드롭하기
x=df1['suicides_no'].values.reshape(-1,1) #2차원으로 변형(자살자 수)
y=df1['gdp_per_capita ($)'].values.reshape(-1,1) #2차원으로 변형(1인당 gdp)
poly_features=PolynomialFeatures(degree=4,include_bias=False) #4차함수로 다항식 변환
x_poly=poly_features.fit_transform(x) #적용시킴
lin_reg=LinearRegression() #선형회귀
lin_reg.fit(x_poly,y) #적용
x_new=np.linspace(min(x),max(x),8364).reshape(8364,1)#등차수열 생성
x_new_poly=poly_features.transform(x_new) #적용해주기
y_predict=lin_reg.predict(x_new_poly) 
plt.plot(x_new,y_predict,'r-') 
plt.plot(x_new,y,'bo') 
plt.xlabel('자살수') 
plt.ylabel('gdp')
plt.title('자살수와 gdp의 연관성') 
plt.show()
```
